```python
movieUrlList =[]
movieReviewUrl = []
i=0
for a in samples:
    if(a['href'].startswith( '/title/' ) and a['href'].endswith( 'adv_li_tt' )):
        movieUrlList.append("https://www.imdb.com"+a['href'])
        movieReviewUrl.append("https://www.imdb.com"+a['href'][:17]+"reviews"+"?ref_=tt_urv")
        print("Found the movie URL:", movieUrlList[i])
        print("Found the moviereview URL:", movieReviewUrl[i])
        i=i+1
```

```python
reviewUrlList = []
for i in range(0,10):
    reviewurl = movieReviewUrl[i]
    print(reviewurl)
    r2 = requests.get(reviewurl) # where url is the above url  
    print("Status code for iteration:",i,r2.status_code)
    c2 = r2.content
    soup2 = BeautifulSoup(c2,"lxml")
    samples2 = soup2.findAll("a", href=True)
    samples2
    j=0
    for a in samples2:
        if(a['href'].startswith( '/review/' ) and a['href'].endswith( 'tt_urv' ) and j<10):
            reviewUrlList.append("https://www.imdb.com"+a['href'])
            print("Found the review URL:",i,reviewUrlList[j])
            j=j+1
        else:
            print('didnt go through')
```
