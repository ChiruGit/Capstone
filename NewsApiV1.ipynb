```python
headers = ["Senator","Source","Author","Title","Published Date"]
pdf = pd.DataFrame(columns=list(headers))# create panda dataframe
```

```python
#for format only
x = 0
for senator in senators: 
    fromDate="2018-01-03"
    toDate="2018-09-03"
    requestUrl = 'https://newsapi.org/v2/everything?q='+senator+'&from='+fromDate+'&to='+toDate+'&apiKey='+mn_key
    #requestUrl = 'https://newsapi.org/v2/everything?q='+senator+'&apiKey=8018a9c44ec54560932db62205b49282' 
    response = requests.get(requestUrl)
    news_content = response.content    
    newsJson = json.loads(news_content)    
    senator_articles = newsJson["articles"]   
    articles_length = len(senator_articles)
    for i in range(0, articles_length):
        source = senator_articles[i]["source"]        
        sourcename = source["name"]
        author = senator_articles[i]["author"] 
        title = senator_articles[i]["title"] 
        pdate = senator_articles[i]["publishedAt"][:10]
        pdf.loc[x]=[senator,sourcename,author,title,pdate]
        x = x + 1
```
