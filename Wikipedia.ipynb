## Wikipedia API
The below block of code uses wikipeidaapi library to extract text from wikipedia pages in text and html formats. It also shows how to extract varoius sections in the wikipage as well as associated URL's listed in the page.

Last block of code shows all the categories, a selected page is associated with.

```python
#Get single page
wiki_wiki = wikipediaapi.Wikipedia('en')

page_py = wiki_wiki.page('Kamala_Harris')
```

```python
#check if page exists
print("Page - Exists: %s" % page_py.exists())
```

```python
#Print page summary
print("Page - Title: %s" % page_py.title)
```

```python
#print page full URL
print(page_py.fullurl)

#print test1 url
#page_py2 = wiki_wiki.page('Test_1')
#print(page_py2.fullurl)
```

```python
#Get full text
wiki_wiki = wikipediaapi.Wikipedia(
        language='en',
        extract_format=wikipediaapi.ExtractFormat.WIKI
)

p_wiki = wiki_wiki.page("Test 1")
print(page_py.text)
```

```python
#Get full text html format
wiki_html = wikipediaapi.Wikipedia(
        language='en',
        extract_format=wikipediaapi.ExtractFormat.HTML
)
p_html = wiki_html.page("Kamala_Harris")
print(p_html.text)
```

```python
#Get page sections
def print_sections(sections, level=0):
        for s in sections:
                print("%s: %s - %s" % ("*" * (level + 1), s.title, s.text[0:40]))
                print_sections(s.sections, level + 1)

print_sections(page_py.sections)
```

```python
#Get links of this page to other pages
def print_links(page):
        links = page.links
        for title in sorted(links.keys()):
            print("%s: %s" % (title, links[title]))

print_links(page_py)
```

```python
#Get page categories
def print_categories(page):
        categories = page.categories
        for title in sorted(categories.keys()):
            print("%s: %s" % (title, categories[title]))


print("Categories")
print_categories(page_py)
```

```python
#Get all pages from a particular category
def print_categorymembers(categorymembers, level=0, max_level=2):
        for c in categorymembers.values():
            print("%s: %s (ns: %d)" % ("*" * (level + 1), c.title, c.ns))
            if c.ns == wikipediaapi.Namespace.CATEGORY and level <= max_level:
                print_categorymembers(c.categorymembers, level + 1)


cat = wiki_wiki.page("Category:American politicians of Indian descent")
print("Category members: Category:American politicians of Indian descent")
print_categorymembers(cat.categorymembers)
```

## Wikipedia and nltk library
The below blocks of code shows the usage of wikipedia and nltk(NLP) librarires to extract information from wikipages and count+show most frequently used words in the page.

```python
#nltk is a good library for natural language processing
import nltk
import wikipedia
```

```python
#In this example, we will get the page for Berkeley, California and count the most commonly used words in the article.
bky = wikipedia.page("Berkeley, California")
bky
```

```python
bk_split = bky.content.split()
```

```python
bk_split[:10]
```

```python
fdist1 = nltk.FreqDist(bk_split)
fdist1.most_common(20)
```
